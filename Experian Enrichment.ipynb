```python
import pandas as pd
import os


# Reading excel first two files provided by Experian and merge into one.

# RPT_NAME_DOB file
data1 = pd.read_csv("RPT_NAME_DOB.csv")

# RPT_ADDRESS file
data2 = pd.read_csv("RPT_ADDRESS.csv")

# Renaming the Customer ID column 
data1=data1.rename(columns=({'dc_unified_id':'CUSTOMERID'}))
data1['CUSTOMERID'] = data1['CUSTOMERID'].astype(str)
data2 = data2.rename(columns=({'dc_unified_id':'CUSTOMERID'}))
data2['CUSTOMERID'] = data2['CUSTOMERID'].astype(str)

# To get the latest date, first reverse the date format  for further comparision.
data2['DATE_REPORTED'] = pd.to_datetime(data2['DATE_REPORTED'] , format='%d/%m/%Y')
# data2.drop('DATE_REPORTED', inplace=True, axis=1)

# To get the latest date, taking the max value.  
data2 = data2[data2.groupby('CUSTOMERID')['DATE_REPORTED'].transform('max') == data2['DATE_REPORTED']]
data2 = data2.drop_duplicates(subset="CUSTOMERID", keep="first", inplace = False)


# Using merge function by setting how='outer' merging the first 2 files
merged_data_1_2 = pd.merge(data1, data2,  
                  on= 'CUSTOMERID', 
                  how='outer')
merged_data_1_2.drop('DATE_REPORTED', inplace=True, axis=1)





# Reading excel 3rd file provided by Experian and merge into the data so far.

# RPT_PHONE file 
data3 = pd.read_csv("RPT_PHONE.csv")

# Renaming the Customer ID column 
data3 = data3.rename(columns=({'dc_unified_id':'CUSTOMERID'}))

#Grouping the rows similar Member reference number.
data3['CUSTOMERID'] = data3['CUSTOMERID'].astype(str)
data3 = data3.groupby('CUSTOMERID').agg(list)

#Removing the columns that are no required in the final output.
data3.drop(['DATE_REPORTED'], inplace=True, axis=1)

#Getting the data in set format from different rows for individual columns.
data3['PHONE'] = data3.PHONE.apply(lambda phone_list: set([phone for phone in phone_list if phone != -99]))

data3 = data3.reset_index(drop=False)

```

```python

# Creating Multiple columns and deleting the previous column 
data3[['Phone 1','Phone 2','Phone 3']] = data3.PHONE.apply(lambda x: pd.Series(str(x).split(",")))
data3.drop('PHONE', inplace=True, axis=1)

# Using merge function by setting how='outer' merging the first 3 files
merged_data_1_3 = pd.merge(merged_data_1_2, data3,
                  on= 'CUSTOMERID', 
                  how='outer')
```

```python
# RPT_EMAIL file 
data4 = pd.read_csv("RPT_EMAIL.csv")

```
